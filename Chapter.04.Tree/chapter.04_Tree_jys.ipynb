## 4.1 트리
### 4.1.1 개요
- 나무를 닮은 자료구조

### 4.1.2 구성요소
- 뿌리(Root) : 가장 위에 있는 노드
- 가지(Branch) : 루트와 잎 사이의 노드
- 잎(Leaf) : 가장 끝의 노드

### 4.1.3 관계
- 부모(Parent)
- 자식(Children)
- 형제(Sibling)

### 4.1.4 경로
- 특정 노드에서 특정 노드를 찾아가기위해 경유하는 노드 리스트
- 길이(Length) : 경로의 노드 개수
- 깊이(Depth) : 루트노드에서 해당 노드까지의 경로의 길이 (루트의 깊이는 0)

### 4.1.5 기타
- 레벨(Level) : 깊이가 같은 노드의 집합
- 높이(Height) : 가장 깊은곳에 있는 잎 노드까지의 깊이
- 노드의 차수(Degree) : 그 노드의 자식 노드 개수
- 트리의 차수(Degree) : 트리 내에 있는 노드들 가운데 자식 노드가 가장 많은 노드의 차수

### 4.1.6 트리 표현하기
- 중첩된 괄호(Nested Parenthesis)
- 중첩된 집합(Nested Set)
- 들여쓰기(Indentation)

### 4.1.7 노드 표현하기
- N-링크 : 노드의 차수가 n개라면 n개의 링크를 가지고 있어야하는 단점 (동적)
- 왼쪽 자식-오른쪽 형제(Left Child-Right Sibling) : 왼쪽자식에서 오른쪽 형제로만 이동

### 4.1.8 트리 구현하기
- 자식 삽입
- 형제 삽입
- 출력하기
- 특정 레벨 모든 노드 출력

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left_child = None
        self.right_sibling = None
        
class LCRSTree:
    def __init__(self, root):
        self.root = root
        
    def appendChild(self, parent, child):
        parent.left_child = child
    
    def appendSibling(self, left, right):
        left.right_sibling = right

    def lcrs(self, node, depth, level=None):
        space = ''
        for i in range(0, depth):
            space += ' '
        if level is not None:
            if level == depth:
                print('%s%s' % (space, node.data))            
        else:
            print('%s%s' % (space, node.data))
        
        if node.left_child is not None:
            self.lcrs(node.left_child, depth + 1, level)
            
        if node.right_sibling is not None:
            self.lcrs(node.right_sibling, depth, level)
        
    def print(self):
        self.lcrs(self.root, 0)
        
    def print_in_level(self, level):
        self.lcrs(self.root, 0, level)

    
root = Node('A')
node_b = Node('B')
node_c = Node('C')
node_d = Node('D')
node_e = Node('E')
node_f = Node('F')
node_g = Node('G')
node_h = Node('H')
node_i = Node('I')
tree = LCRSTree(root)
tree.appendChild(root, node_b)
tree.appendSibling(node_b, node_c)
tree.appendChild(node_b, node_d)
tree.appendSibling(node_c, node_e)
tree.appendChild(node_e, node_f)
tree.appendChild(node_f, node_g)
tree.appendChild(node_g, node_h)
tree.appendSibling(node_g, node_i)
print('print()')
tree.print()
print('')
print('print_in_level()')
tree.print_in_level(3)
```

## 4.2 이진트리 (Binary Tree)

### 4.2.1 개요
- 모든 노드가 최대 2개의 자식을 가질 수 있는 트리
- 노드의 최대 차수가 2개
- 수식 이진 트리(Expression Binary Tree) : 수식을 트리 형태로 표현하여 계산가능
- 이진 탐색 트리(Binary Search Tree) : 아주 빠른 데이터 검색가능

### 4.2.2 형태
- 포화 이진 트리(Full Binary Tree) : 잎 노드들이 모두 같은 깊이에 존재
- 완전 이진 트리(Complete Binary Tree) : 잎 노드들이 트리의 왼쪽부터 차곡차곡 채워진 형태
- 높이 균형 트리(Height Balanced Tree) : 루트 노드를 기준으로 왼쪽 하위트리와 오른쪽 하위트리의 높이가 1이상 차이나지 않는 이진 트리
- 완전 높이 균형 트리(Completely Height Binary Tree) : 루트 노드를 기준으로 왼쪽 하위트리와 오른쪽 하위트리의 높이가 같은 이진 트리

### 4.2.3 순회
- 트리 내의 노드들 사이를 돌아다니는 것

### 4.2.4 전위 순회
- 루트 노드 -> 왼쪽 하위 트리 -> 오른쪽 하위 트리

### 4.2.5 중위 순회
- 왼쪽 하위 트리 -> 루트 노드 -> 오른쪽 하위 트리

### 4.2.6 후위 순회
- 왼쪽 하위 트리 -> 오른쪽 하위 트리 -> 루트 노드

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = str(data)
        
class SBT_Tree:
    def __init__(self, root):
        self.root = root

    def appendChild(self, parent, child):
        if parent.left is None:
            parent.left = child
        elif parent.right is None:
            parent.right = child
    
    def print_postorder(self):
        print('print_postorder')
        print(self.make_postorder_expression(self.root, ''))
    
    def print_preorder(self):
        print('print_preorder')
        print(self.make_preorder_expression(self.root, ''))
    
    def print_order(self):
        print('print_order')
        print(self.make_order_expression(self.root, ''))
        
    def make_order_expression(self, node, expression):
        if node.left is not None:
            expression = self.make_order_expression(node.left, expression)
            expression += node.data
            expression = self.make_order_expression(node.right, expression)
        else:
            expression += node.data
        return expression
            
    def make_preorder_expression(self, node, expression):
        expression += node.data
        if node.left is not None:
            expression = self.make_preorder_expression(node.left, expression)
            expression = self.make_preorder_expression(node.right, expression)
        return expression
            
    def make_postorder_expression(self, node, expression):
        
        if node.left is not None:
            expression = self.make_postorder_expression(node.left, expression)
            expression = self.make_postorder_expression(node.right, expression)
            expression += node.data
        else:
            expression += node.data
        return expression
            
        
# (1*2)+(7-8)
node1 = Node(1)
node2 = Node('*')
node3 = Node(2)
node4 = Node('+')
node5 = Node(7)
node6 = Node('-')
node7 = Node(8)
tree = SBT_Tree(node4)
tree.appendChild(node4, node2)
tree.appendChild(node4, node6)
tree.appendChild(node2, node1)
tree.appendChild(node2, node3)
tree.appendChild(node6, node5)
tree.appendChild(node6, node7)

# (1*2)+(7-8)
print('')
tree.print_order()

# +*12-78
print('')
tree.print_preorder()

# 12*78-+
print('')
tree.print_postorder()
```

## 4.3 수식 트리 (Expression Tree)

### 4.3.1 개요
- 수식을 표현하는 이진 트리 (Expression Binary Tree)
- 피연산자는 잎 노드
- 연산자는 루트 노드, 또는 가지 노드

### 4.3.2 구축
- 후회 표기식을 입력받는다 가정
- 수식을 뒤에서부터 앞쪽으로 읽어나감
- 수식에서 제일 마지막에 있는 토큰은 루트 노드, 항상 연산자
- 수식에서 읽어낸 토큰이 연산자인 경우 가지노드, 뒤따라 오는 토큰은 각각 오른쪽, 왼쪽 자식 노드
- 다음 토큰에도 연속해서 연산자인 경우, 이 토큰으로부터 만들어지는 하위 트리가 완성된 이후에 읽어낸 토큰이 왼쪽 자식 노드
- 수식에서 읽어낸 토큰이 숫자이면 잎노드

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = str(data)

class EBT_Tree:
    root = None
    
    def __init__(self, postorder_str):
        postorder_arr = list(postorder_str)
        self.root = Node(postorder_arr.pop())
        self.make_tree(postorder_arr, self.root)
        
    def make_tree(self, arr, node):
        if len(arr) <= 0:
            return
            
        current = arr.pop()
        if current.isdigit():
            right_node = Node(current)
            left_node = Node(arr.pop())
            node.right = right_node
            node.left = left_node
        else:
            if node.right is None:
                right_node = Node(current)
                node.right = right_node
                if not arr[len(arr)-1].isdigit():
                    node.left = Node(arr.pop())
                self.make_tree(arr, node.right)
            if node.left is None:
                node.left = Node(arr.pop())
                self.make_tree(arr, node.left)

    def calculate(self):
        expression = self.make_postorder_expression(self.root, '')
        print(expression[1])
        
    def make_postorder_expression(self, node, expression):
        result = 0
        if node.left is not None:
            expression, left_result = self.make_postorder_expression(node.left, expression)
            expression, right_result = self.make_postorder_expression(node.right, expression)
            
            if len(expression) <= 0:
                expression = str(left_result) + '|' + str(right_result)
            result = self.calc_expression(expression, node.data)
            expression = ''
        else:
            expression += node.data
        return expression, result
    
    def calc_expression(self, digit_str, expression):
        if len(digit_str) > 2:
            digit_list = [float(s) for s in digit_str.split('|')]
            left_digit = digit_list[0]
            right_digit = digit_list[1]
        else:
            left_digit = float(digit_str[0])
            right_digit = float(digit_str[1])
        if expression == '*':
            result = left_digit * right_digit
        elif expression == '+':
            result = left_digit + right_digit
        elif expression == '-':
            result = left_digit - right_digit
        elif expression == '/':
            result = left_digit / right_digit
        return result
    
tree = EBT_Tree('12*78-+')
print('12*78-+ : 1')
tree.calculate()

tree = EBT_Tree('63/25+*')
print('63/25+* : 14')
tree.calculate()
```

## 4.4 분리 집합 (Disjoint Set)

### 4.4.1 개요
- 교집합을 갖지 않는 집합들
- 원소 또는 개체가 "어느 집합에 소속되어 있는가?" 라는 정보를 바탕으로 하는 알고리즘에 응용

### 4.4.2 표현
- 자식이 부모를 가리킴
- 루트 노드는 집합 그 자체, 그 자식 노드들이 모두 루트 노드를 가리킴

### 4.4.3 연산
- 합집합 (Union)
- 집합 탐색 (Find) : 원소가 속해 있는 집합을 찾는 연산, 부모가 Null 인 루트 노드를 찾는것과 같음

```python
class Node:
    def __init__(self, data):
        self.parent = None
        self.data = str(data)
    
    def __str__(self):
        return self.data
        
class DisjointSet:
    def __init__(self, root):
        self.root = root
    
    def union(self, dis_set):
        dis_set.root.parent = self.root
    
    def find(self, target_node):
        current = target_node
        while(current.parent is not None):
            current = current.parent
        return current
    
    def make(self, nodes):
        for node in nodes:
            node.parent = self.root
        
root_a = Node('ROOT_A')
root_a_nodes = [Node(i) for i in range(1, 5)]
root_b = Node('ROOT_B')
root_b_nodes = [Node(i) for i in range(6, 10)]

dis_set_a = DisjointSet(root_a)
dis_set_a.make(root_a_nodes)

dis_set_b = DisjointSet(root_b)
dis_set_b.make(root_b_nodes)

print('find 1 = ROOT_A')
print(dis_set_a.find(root_a_nodes[0]))
print('')

print('find 6 = ROOT_B')
print(dis_set_b.find(root_b_nodes[0]))
print('')

print('union a b')
dis_set_a.union(dis_set_b)

print('find 6 = ROOT_A')
print(dis_set_a.find(root_b_nodes[0]))
```
